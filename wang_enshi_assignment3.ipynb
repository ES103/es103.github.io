```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("building_inventory.csv")
```

```python
df_clear1 = df.drop(df[(df['Year Acquired']==0) | (df['Year Constructed']==0)].index)
plt.plot(df_clear1["Year Constructed"],df_clear1["Year Acquired"],'.')
plt.xlabel("Year Constructed")
plt.ylabel("Year Acquired")
```

Q1:Relationship between the year acquired and the year constructed

I used pandas to read the CSV file and then eliminated the invalid values for the x and y variables, which are the 0 values in the year constructed and the year acquired. Then I plotted the graph with year constructed and x-axis and year acquired as the y-axis. All the things work for
me. 

The success is that I can see a positive linear relationship between year constructed and year acquired. However, we may not quantify the outliers that we can see in the graph.   

```python
df_clear2 = df.drop(df[df['Square Footage']==0].index)
df_by_congdist = df_clear2.groupby("Congress Dist")
total_sqrt = df_by_congdist["Square Footage"].sum()
plt.xticks(np.arange(0,19,1))
plt.bar(total_sqrt.index, total_sqrt)
plt.ylabel("Total Square Footage")
plt.xlabel("Congress District")
```

Q2:Total square footage as a function of congressional district ("Congress Dist")

I used the data imported before and then eliminate the invalid values for the x variables,
which is the 0 values in square footage. After that, I group the data by the congressional district and sum the values. At last, I plotted the sum values as the y-axis and the index of
congress district as the x-axis. I found that the x-axis is not in integer so I used xticks to make it be integers.

The success is that I see the total square footage is in the range of 0-1.5 but extremely high in the district of 13. The shortcoming is that the total square footage seems to have a period of 5 but this graph cannot prove it. 


```python
average = df_by_congdist["Square Footage"].sum()/df_by_congdist["Total Floors"].sum()
plt.xticks(np.arange(0,19,1))
plt.bar(average.index, average)
plt.ylabel("Average Square Footage")
plt.xlabel("Congress District")
```

Q3:Average square footage per floor as a function of congressional district

I used the data used before and then calculated the average square footage. Also, the x-axis
should be integers and I used the xticks to do it. I chose the bar chart here since a congressional district is a discrete number and the bar chart is suitable to describe this kind of data.

I found that the first several districts tend to have higher average square footage and the last several districts tend to have less. The shortcoming is that this tendency can not be qualified or
